### VERIFICANDO VALORES VAZIOS

#### BASE COMPLETA

```python
# Inicializar um dicionário para armazenar o número de valores vazios em cada coluna
num_valores_vazios = {}

# Loop sobre cada coluna do Df
for coluna in df.columns:
    # Contar o número de valores vazios na coluna atual e armazenar no dicionário
    num_valores_vazios[coluna] = df[coluna].isnull().sum()

# Imprimir o número de valores vazios em cada coluna
for coluna, num_vazios in num_valores_vazios.items():
    print(f"A coluna {coluna} possui {num_vazios} valores vazios.")
    
```

#### BASE DEPRESSÃO ENTRE 20 E 40 ANOS

```python
import time

# Registrar o tempo inicial
inicio = time.time()


# Inicializar um dicionário para armazenar o número de valores vazios em cada coluna
num_valores_vazios = {}

# Loop sobre cada coluna do Df
for coluna in df_depre_filtrado.columns:
    # Contar o número de valores vazios na coluna atual e armazenar no dicionário
    num_valores_vazios[coluna] = df_depre_filtrado[coluna].isnull().sum()

# Imprimir o número de valores vazios em cada coluna
for coluna, num_vazios in num_valores_vazios.items():
    print(f"A coluna {coluna} possui {num_vazios} valores vazios.")

# Registrar o tempo final
fim = time.time()

# Calcular o tempo de execução
tempo_execucao = fim - inicio

# Imprimir o tempo de execução
print(f"Tempo de execução: {tempo_execucao:.2f} segundos")
```
